## Research question/interests

Everyone in the world will somehow spend money on medical things, which makes me want to know the cost for different people. So here is my question, how much more money will a male smoker aged over 50 spend than an adult male smoker?

```python
import pandas as pd
```

```python
data = pd.read_csv("../data/raw/Medical_Cost.csv")
```

```python
data
```

```python
data2 = data.drop(["bmi", "children", "region"], axis = "columns")
```

```python
Over50_data = data2[(data2['age'] > 50) & (data['sex'] == 'male') & (data['smoker'] == 'yes')]
```

```python
Over50_Final = Over50_data.sort_values('age', ascending = False)
```

```python
Over50_Final
```

```python
Over18_data = data2[(data2['age'] > 18) & (data2['age'] < 50) & (data['sex'] == 'male') & (data['smoker'] == 'yes')]
Over18_Final = Over18_data.sort_values('age', ascending = False)
Over18_Final
```
